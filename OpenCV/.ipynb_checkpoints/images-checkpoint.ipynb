#### Write an image

```python
# program loads an image in grayscale, displays it, 
#save the image if you press ‘s’ and exit, or simply
#exit without saving if you press ESC key
import numpy as np
import cv2

img = cv2.imread('Data/messi5.jpg',0)
cv2.imshow('image',img)
k = cv2.waitKey(0)
if k == 27:         # wait for ESC key to exit
    cv2.destroyAllWindows()
elif k == ord('s'): # wait for 's' key to save and exit
    cv2.imwrite('messigray.png',img)
    cv2.destroyAllWindows()
```

#### Using Matplotlib

```python
#You can zoom images, save it etc using Matplotlib.
import cv2
from matplotlib import pyplot as plt

img = cv2.imread('Data\messi5.jpg',0)
plt.imshow(img, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```

### OpenCV follows BGR order, while matplotlib likely follows RGB order.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('Data\messi5.jpg')
b,g,r = cv2.split(img)

img2 = cv2.merge([r,g,b])

plt.subplot(121);plt.imshow(img) # expects distorted color
plt.subplot(122);plt.imshow(img2) # expect true color
plt.show()

cv2.imshow('bgr image',img) # expects true color
cv2.imshow('rgb image',img2) # expects distorted color
cv2.waitKey(0)
cv2.destroyAllWindows()
```
